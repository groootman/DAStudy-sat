```python
print("----------------------------------------------------------------------------")

for column in data.columns:
    if len(data[column].unique()) > 10:
        print(column + " unique값 확인")
        print(data[column].unique()[:10], "...")
        print(len(data[column].unique()), "개")
        print("----------------------------------------------------------------------------")
    else:
        print(column + " unique값 확인")
        print(data[column].unique())
        print(len(data[column].unique()), "개")
        print("----------------------------------------------------------------------------")
```

```python
print("---------------------------")

row_count = data.shape[0]

for column in data.columns:
    null_count = data[column].isnull().sum()
    if null_count > 0:
        print(column + " 결측치 비율")
        print(round(null_count / row_count * 100, 2), "%")
        print("---------------------------")
    else:
        pass
```

```python
data_copy_1 = data

group_data = data_copy_1.groupby(["user_session", "event_type"]).agg(datetime = ("event_time", "min"))

temp_df = pd.DataFrame(data_copy_1['event_type'].unique(), columns = ["event_type"])
temp_df["step"] = [1, 2, 3]
temp_df.set_index("event_type", append = False, inplace = True)

merge_data = pd.merge(group_data, temp_df, left_on = "event_type", right_index = True)

pivot_data = merge_data.reset_index().pivot(index = ["user_session"], columns = "step", values = "datetime")

pivot_data.columns = temp_df.index
pivot_data
```

```python
step_count = [pivot_data[event].notnull().sum() for event in pivot_data.columns]

print("event별 횟수")
for num, event in enumerate(step_count):
    print(pivot_data.columns[num], " : ", event, "회")
    
print("")

print("event별 전환율")
for num, event in enumerate(step_count[1:]):
    step_rate = round(event / step_count[num] * 100, 2)
    print(pivot_data.columns[num+1], " 전환율 : ", step_rate, "%")
```

```python
data_copy_2 = data

# 결측값 제거
data = data_copy_2.drop("user_id", axis = 1)
data = data_copy_2.dropna(how='any')

# session별 / category별 / brand별 funnel 확인
group_data = data_copy_2.groupby(["user_session", "category_code", "brand", "event_type"]).agg(datetime = ("event_time", "min"))

temp_df = pd.DataFrame(data_copy_2['event_type'].unique(), columns = ["event_type"])
temp_df["step"] = [1, 2, 3]
temp_df.set_index("event_type", append = False, inplace = True)

merge_data = pd.merge(group_data, temp_df, left_on = "event_type", right_index = True)

pivot_data = merge_data.reset_index().pivot(index = ["user_session", "category_code", "brand"], columns = "step", values = "datetime")

pivot_data.columns = temp_df.index
pivot_data
```

```python
purchase_data = pivot_data[pivot_data["purchase"].notnull() == True].dropna(how = "any")
purchase_data
```

```python
purchase = purchase_data.reset_index()
purchase_group = purchase.groupby(["category_code", "brand"]).count().reset_index()
purchase_group = purchase_group.drop(purchase_group.columns[2:5], axis = 1)

best_brand = pd.DataFrame(columns = ["category_code", "brand", "purchase"])

for category in purchase_group["category_code"].unique():
    
    temp_1 = purchase_group[purchase_group["category_code"] == category]
    temp_2 = temp_1[temp_1["purchase"] == temp_1["purchase"].max()]
    best_brand = pd.concat((best_brand, temp_2), ignore_index = True)
    
best_brand
```
